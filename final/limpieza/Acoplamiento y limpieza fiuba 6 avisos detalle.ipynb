### Mergeo 2 conjuntos de csv

```python
notice_aux_1 = pd.merge(notice_details_from_15, notice_details_from_15_missing, on = None, how = 'outer')
notice_aux_2 = pd.merge(notice_details_original, notice_details_to_15, on = None, how = 'outer')
```

### Mergeo los dos conjuntos de csv

```python
notice_details = pd.merge(notice_aux_1, notice_aux_2, on = None, how = 'outer')
notice_details.head()
```

```python
print(len(notice_details))
```

### Hago una limpieza de la descripción, eliminando el código HTML

```python
cleanr = re.compile('<.*?>')
def cleanhtml(raw_html):
    cleantext = re.sub(cleanr, '', raw_html)
    return cleantext
```

```python
notice_details["descripcion"] = notice_details["descripcion"].apply(lambda x: cleanhtml(x))
```
