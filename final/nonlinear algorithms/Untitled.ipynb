```python
# Load libraries
import pandas as pd
import numpy as np
from sklearn import model_selection
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
training = pd.read_csv('../csvs/training.csv')
```

```python
training.drop('sepostulo', 1, inplace = True)
```

```python
X = training.drop('nombre_area', 1)
Y = training['nombre_area']
```

```python
X
```

```python
Y
```

```python
# Splitting the dataset
split_percentage = 0.8
split = int(split_percentage*len(training))
X_train = X[:split]
Y_train = Y[:split]
X_val = X[split:]
Y_val = Y[split:]
```

```python
knn = KNeighborsClassifier(n_neighbors = 80)
knn.fit(X_train, Y_train)

accuracy_train = accuracy_score(Y_train, knn.predict(X_train))
accuracy_val = accuracy_score(Y_val, knn.predict(X_val))

print ('Train data accuracy: %.2f' % accuracy_train)
print ('Val data accuracy: %.2f' % accuracy_val)
```
